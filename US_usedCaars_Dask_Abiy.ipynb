## Four selected parameters affecting the price of used cars
- mileage, days_on_market, city_fuel_economy and year

```python
df_price=cars1[['price','mileage','city_fuel_economy','year','daysonmarket']]
df_price=df_price.sample(frac=0.1,random_state=2)
df_price.compute()
```

```python

p_mil = figure(plot_width=250, plot_height=250, title="Mileage", x_axis_label='Mileage', y_axis_label='Price')
p_mil.scatter(df_price.mileage, df_price.price, size=2, color="blue")

p_days = figure(plot_width=250, plot_height=250, title="Days_on_ market", x_axis_label='Days_on_market', y_axis_label='Price')
p_days.scatter(df_price.daysonmarket, df_price.price, size=2, color="red")

p_fuel = figure(plot_width=250, plot_height=250, title="Fuel economy", x_axis_label='Fuel_economy', y_axis_label='Price')
p_fuel.scatter(df_price.city_fuel_economy,df_price.price, size=2,color="green")

p_year = figure(plot_width=250, plot_height=250, title="Manufacture year", x_axis_label='Year of manufacture', y_axis_label='Price')
p_year.scatter(df_price.year, df_price.price, size=2,color="purple")

p=gridplot(children=[[p_mil,p_days],[p_fuel,p_year]],sizing_mode = 'stretch_both')



output_notebook()
show(p)
```

# Multiple linear regression for price prediction by considering the four parameters

- Multiple linear regression 

$Price=k+\alpha*M_i+\beta*D_i+\gamma*F_i+\mu*Y_i+e$

- $\alpha$ -mileage index
- $\beta$ - days on market index
- $\gamma$ - fuel economy index
- $\mu$ - year of manufacture index
- $e$ - error

```python
x = df_price[['mileage', 'city_fuel_economy', 'daysonmarket','year']]
y = df_price['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
import dask_ml
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test =dask_ml.model_selection.train_test_split(x, y, 
                                                                           test_size = 0.3, random_state = 100)
#------->70% train data and 30% test data
```

```python
mlr = LinearRegression()  
mlr.fit(x_train, y_train)
```

```python
print("K =", mlr.intercept_)
print("Coefficients:")
#coeff_li=['$\alpha$','F_i','D_i','Y_i']
list(zip(x, mlr.coef_))
```

```python
print('R squared: {:.2f}'.format(mlr.score(x_test,y_test)*100))
```
